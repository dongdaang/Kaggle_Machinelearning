```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 200)
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
train_data.shape, test_data.shape
```

```python
train_test_data = pd.concat([train_data, test_data])
```

```python
train_test_data.shape
```

```python
object_columns = train_test_data.dtypes[train_test_data.dtypes == 'object'].index.tolist()
object_columns
```

```python
for column in object_columns:
    train_test_data[column] = pd.factorize(train_test_data[column])[0]
```

```python
train_test_data.info()
```

```python
train_test_data = train_test_data.fillna(-999)
```

```python
train_test_data.isnull().sum()
```

```python
train_data = train_test_data[train_test_data['SalePrice'] != -999]
test_data = train_test_data[train_test_data['SalePrice'] == -999]
train_data.shape, test_data.shape
```

```python
test_data = test_data.drop('SalePrice', axis=1)
```

```python
test_data.shape
```

```python
train_data['SalePrice'].max(), train_data['SalePrice'].min()
```

```python
SaleBand = train_data['SalePrice'].tolist()
SaleBand
```

```python
for i in range(len(SaleBand)):
    if SaleBand[i] > 34900 and SaleBand[i] <= 100000:
        SaleBand[i] = 0
    elif SaleBand[i] > 100000 and SaleBand[i] <= 200000:
        SaleBand[i] = 1
    elif SaleBand[i] > 200000 and SaleBand[i] <= 300000:
        SaleBand[i] = 2
    else:
        SaleBand[i] = 3
        
SaleBand
```

```python
train_data['SalePrice'] = SaleBand
```

```python
train_data.head(10)
```

```python
ftr_train = train_data.drop(['Id', 'SalePrice'], axis=1)
target_train = train_data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split

train_x, valid_x, train_y, valid_y = train_test_split(ftr_train, target_train, test_size=0.3, random_state=2022)
train_x.shape, valid_x.shape
```

```python
from lightgbm import LGBMClassifier

clf = LGBMClassifier(
    n_jobs=-1,
    n_estimators=1000,
    learning_rate=0.02,
    num_leaves=16,
    subsample=0.8,
    max_depth=8,
    silent=-1,
    verbose=-1
)

clf.fit(train_x, train_y, eval_set=[(train_x, train_y), (valid_x, valid_y)],
       eval_metric='multi_logloss', verbose=50, early_stopping_rounds=100)
```

```python
from lightgbm import plot_importance

plot_importance(clf, figsize=(16, 32))
```
