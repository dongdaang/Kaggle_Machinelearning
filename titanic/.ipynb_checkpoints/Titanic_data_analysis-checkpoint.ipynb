```python
def show_column_hist_by_survived(df, columns):
    cond1 = (df['Survived'] == 1)
    cond0 = (df['Survived'] == 0)
    
    for column in columns:
        fig, axs = plt.subplots(figsize=(12, 4), nrows=1, ncols=2, squeeze=False)
        sns.violinplot(x='Survived', y=column, data=df, ax=axs[0][0])
        sns.distplot(df[cond1][column], label='1', color='red', ax=axs[0][1])
        sns.distplot(df[cond0][column], label='0', color='blue', ax=axs[0][1])
```

```python
columns_continuous_value = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
```

```python
show_column_hist_by_survived(train_data, columns_continuous_value)
```

```python
columns_object = train_data.dtypes[train_data.dtypes == 'object'].index.tolist()
columns_object
```

```python
def show_category_by_survived(df, columns):
    for column in columns:
        print('column name:', column)
        chart = sns.catplot(x=column, col='Survived', data=df, kind='count')
        chart.set_xticklabels(rotation=65)
```

```python
show_category_by_survived(train_data, columns_object)
```

```python
critical_columns = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']

col_corr = train_data[critical_columns].corr()
col_corr
```

```python
plt.figure(figsize=(9, 9))
sns.heatmap(col_corr, annot=True)
```
